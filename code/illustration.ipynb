```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the continuous signal (e.g., a sine wave)
def continuous_signal(t):
    return (np.sin(2 * np.pi * 0.5 * t) + 0.5 * np.sin(2 * np.pi * 1.5 * t))*np.exp(-0.5*t)

# Continuous time domain for smooth plot
t_continuous = np.linspace(0, 4, 1000)  # Smooth time axis (1000 points)
y_continuous = continuous_signal(t_continuous)

# Discrete time domain (sampling points)
num_samples = 16  # Number of sampled points
t_discrete = np.linspace(0, 4, num_samples)  # Sampling points
y_discrete = continuous_signal(t_discrete)

# Plot the continuous signal
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(t_continuous, y_continuous, label="Sinal contínuo", 
         color="#d7e1ee", linewidth=2, linestyle='--')

# Plot the sampled points
ax.scatter(t_discrete, y_discrete, color="#991f17", label="Amostragem", zorder=5)
ax.stem(t_discrete, y_discrete, basefmt=" ", linefmt="#df8879", markerfmt="ro", label="Sinal discreto")

# Customize the plot
plt.title("Illustração de um sinal discreto", fontsize=14)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
plt.xlabel("Time", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```
